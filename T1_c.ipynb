```python
# Name - Dev Patel
# Roll No - 18110113



import numpy as np

links = 3
# Some constants
L1 = 0.25
L2 = 0.25
L3 = 0.25

# User input Coordinates
Coordinates = np.array([])
for i in range(1,5):
    if(i==1):
        Co = list(map(float, input("Coordinates of point {}:".format(i)).split()))
        Coordinates = np.array(Co).reshape(1,3)
    else:
        Co = list(map(float, input("Coordinates of point {}:".format(i)).split()))
        Coordinates = np.r_[Coordinates,np.array(Co).reshape(1,3)]

A = np.array(Coordinates[0,:])
B = np.array(Coordinates[1,:])
C = np.array(Coordinates[2,:])
D = np.array(Coordinates[3,:])

#STANFORD
def inverse_kinematics_Stanford(O):

    r = (O[0]**2 + O[1]**2)**0.5
    s = O[2] - L1
    theta1 = np.arctan2(O[1],O[0]) 
    d3 = (r**2 + s**2  - L2**2)**0.5
    theta2 = -np.arctan2(-s,r) + np.arctan2(d3,L2)
    

    I = np.array([theta1,theta2,d3])
    return(I)
# Forward Kinematics 

def check_end_Effector_position_Stanford(I):
    T = np.eye(4)
    DH_parameters = np.array([[I[0],0,L1,1.57],[I[1],L2,0,1.57],[0,0,I[2],0]])
    for i in range(3):
        A = np.array([[np.cos(DH_parameters[i,0]), -np.sin(DH_parameters[i,0])*np.cos(DH_parameters[i,3]), np.sin(DH_parameters[i,0])*np.sin(DH_parameters[i,3]),DH_parameters[i,1]*np.cos(DH_parameters[i,0])],\
                [np.sin(DH_parameters[i,0]), np.cos(DH_parameters[i,0])*np.cos(DH_parameters[i,3]), -np.cos(DH_parameters[i,0])*np.sin(DH_parameters[i,3]),DH_parameters[i,1]*np.sin(DH_parameters[i,0])],\
                    [0,np.sin(DH_parameters[i,3]),np.cos(DH_parameters[i,3]),DH_parameters[i,2]],\
                        [0,0,0,1]])

        T = np.dot(T,A)
    return(print('x=',T[0,3],'y=',T[1,3],'z=',T[2,3]))

input = inverse_kinematics_Stanford(A)
print(" For point A \n Theta1 = ", input[0], "\n Theta2 =", input[1],"\n Extension: ", input[2], "\n")
Check = print(check_end_Effector_position_Stanford(input))

input = inverse_kinematics_Stanford(B)
print(" For point B \n Theta1 = ", input[0], "\n Theta2 =", input[1],"\n Extension: ", input[2], "\n")
Check = print(check_end_Effector_position_Stanford(input))

input = inverse_kinematics_Stanford(C)
print(" For point C \n Theta1 = ", input[0], "\n Theta2 =", input[1],"\n Extension: ", input[2], "\n")
Check = print(check_end_Effector_position_Stanford(input))

input = inverse_kinematics_Stanford(D)
print(" For point D \n Theta1 = ", input[0], "\n Theta2 =", input[1],"\n Extension: ", input[2], "\n")
Check = print(check_end_Effector_position_Stanford(input))
```

```python
# SCARA manipulator
d1 = 0.25
# Some constants
L1 = 0.25
L2 = 0.25
L3 = 0.25
def inverse_kinematics_SCARA(O):

    c2 = (O[0]**2 + O[1]**2 - L1**2 - L2**2)/(2*L1*L2)
    theta2 = np.arctan2(-(1-c2**2)**0.5,c2)
    theta1 = np.arctan2(O[1],O[0]) - np.arctan2(L2*np.sin(theta2),L1 + L2*np.cos(theta2))

    d3 = d1 - O[2] 

    I = np.array([theta1,theta2,d3])
    return(I)

# gets the value of end effector position for the given joint variables
def check_end_Effector_position_SCARA(I):
    T = np.eye(4)
    DH_parameters = np.array([[0,0,d1,0],[I[0],L1,0,0],[I[1],L1,0,3.14],[0,0,I[2],0]])
    for i in range(4):
        A = np.array([[np.cos(DH_parameters[i,0]), -np.sin(DH_parameters[i,0])*np.cos(DH_parameters[i,3]), np.sin(DH_parameters[i,0])*np.sin(DH_parameters[i,3]),DH_parameters[i,1]*np.cos(DH_parameters[i,0])],\
                [np.sin(DH_parameters[i,0]), np.cos(DH_parameters[i,0])*np.cos(DH_parameters[i,3]), -np.cos(DH_parameters[i,0])*np.sin(DH_parameters[i,3]),DH_parameters[i,1]*np.sin(DH_parameters[i,0])],\
                    [0,np.sin(DH_parameters[i,3]),np.cos(DH_parameters[i,3]),DH_parameters[i,2]],\
                        [0,0,0,1]])

        T = np.dot(T,A)
    return(print('x=',T[0,3],'y=',T[1,3],'z=',T[2,3]))

input = inverse_kinematics_SCARA(A)
print(" For point A \n Theta1 = ", input[0], "\n Theta2 =", input[1],"\n Extension: ", input[2], "\n")
Check = print(check_end_Effector_position_SCARA(input))

input = inverse_kinematics_SCARA(B)
print(" For point B \n Theta1 = ", input[0], "\n Theta2 =", input[1],"\n Extension: ", input[2], "\n")
Check = print(check_end_Effector_position_SCARA(input))

input = inverse_kinematics_SCARA(C)
print(" For point C \n Theta1 = ", input[0], "\n Theta2 =", input[1],"\n Extension: ", input[2], "\n")
Check = print(check_end_Effector_position_SCARA(input))

input = inverse_kinematics_SCARA(D)
print(" For point D \n Theta1 = ", input[0], "\n Theta2 =", input[1],"\n Extension: ", input[2], "\n")
Check = print(check_end_Effector_position_SCARA(input))
```
