```python
# Scara Inverse and forward Kinematics

def scara(ee_pos,links):

    #inverse Kinematics SCARA

    r = abs((ee_pos[0]**2+ee_pos[1]**2-links[0]**2-links[1]**2)/(2*links[0]*links[1]))
   
    theta2 = np.arctan(np.sqrt(abs(1-r**2))/r)
    theta1 = np.arctan(ee_pos[1]/ee_pos[0]) - np.arctan((links[1]*np.sin(theta2))/(links[0]+links[1]*np.cos(theta2)))
   
    d3 = links[0]-ee_pos[2]
    
    print("SCARA Inv Results: \n", "theta1 = ", theta1, "\n theta2 =", theta2, "\n Extension: ", d3, "\n")
    
    #forward kinematics SCARA

    x=links[0]*np.cos(theta1)+links[1]*np.cos(theta1+theta2)
    y=links[0]*np.sin(theta1)+links[1]*np.sin(theta1+theta2)
    z=links[2]-d3

    print("SCARA for Results: \n", "x = ", x,"\n y =", y,"\n = ", z, "\n")

#inputs
scara([0.45, 0.075, 0.1],[0.25,0.25,0.25])
scara([0.45, -0.075, 0.1],[0.25,0.25,0.25])
scara([0.25, -0.075, 0.1],[0.25,0.25,0.25])
scara([0.25, 0.075, 0.1],[0.25,0.25,0.25])
```

```python
# PUMA Inverse and forward Kinematics

def puma(ee_pos,links):

    #Inverse kinematics

    theta1 = np.arctan(ee_pos[0]/ee_pos[1])
    D = (ee_pos[0]**2+ee_pos[1]**2+(ee_pos[2]-links[0])**2-links[1]**2-links[2]**2)/(2*links[1]*links[2])
  
    theta3 = np.arctan((D/-math.sqrt(1-D*D)))
    a= np.arctan((math.sqrt(ee_pos[0]**2+ee_pos[1]**2))/ee_pos[2]-links[0])
    b= np.arctan((links[1]+links[2]*math.cos(theta3))/(links[2]*math.sin(theta3)))
    theta2=a-b
     
    print("puma Inv kin: \n", "Theta1 = ", theta1 , "\n Theta2 =", theta2, "\n Theta3: ", theta3, "\n")

    #forward kinematics

    x=(links[1]*np.cos(theta2)+links[2]*np.cos(theta2+theta3))*np.cos(theta1)
    y=(links[1]*np.cos(theta2)+links[2]*np.cos(theta2+theta3))*np.sin(theta1)
    z=links[0]+links[1]*np.sin(theta2)+ links[2]*np.sin(theta2+theta3)

    print("puma For kin: \n", "x = ", x, "\n y =", y, "\n z= ", z, "\n")

#inputs
puma([0.45, 0.075, 0.1],[0.25,0.25,0.25])
puma([0.45, -0.075, 0.1],[0.25,0.25,0.25])
puma([0.25, -0.075, 0.1],[0.25,0.25,0.25])
puma([0.25, 0.075, 0.1],[0.25,0.25,0.25])
```

```python
# Stanford Inverse and forward Kinematics


def stanford(ee_pos,links):

    #innverse kinematics

    theta1 = np.arctan(ee_pos[1]/ee_pos[0])
    r = np.sqrt(ee_pos[0]**2 + ee_pos[1]**2)
    s = ee_pos[2] - links[0]
    theta2 = np.arctan(s/r)
    d3 = np.sqrt(r**2 + s**2) - links[1]

    print("Stanford inv res: \n", "theta1 = ", theta1, "\n theta2 =", theta2,"\n Extension= ", d3,"\n")


    #forward kinematics

    x=(links[1]+d3)*np.cos(theta1)*np.cos(theta2)
    y=(links[1]+d3)*np.sin(theta1)*np.cos(theta2)
    z=links[0]+(links[1]+d3)*np.sin(theta2)
    print("Stanford for res: \n", "x = ", x,"\n y =", y,"\n z= ", z, "\n")

#inputs
stanford([0.45, 0.075, 0.1],[0.25,0.25,0.25])
stanford([0.45, -0.075, 0.1],[0.25,0.25,0.25])
stanford([0.25, -0.075, 0.1],[0.25,0.25,0.25])
stanford([0.25, 0.075, 0.1],[0.25,0.25,0.25])
```
