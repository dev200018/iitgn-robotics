```python
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = open('Gait_DATA.csv')
data = csv.reader(file)
header = []
header = next(data)
rows = []
for row in data:
    rows.append(row)
```

```python
newlst=[]
for i in range(len(rows)):
    newlst.append([float(rows[i][0]),float(rows[i][1])])
```

```python
temp=np.transpose(newlst)
x=temp[0]
y=temp[1]
```

```python
plt.plot(x,y)
plt.xlabel('x - axis (in cm)')
plt.ylabel('y - axis (in cm)')
plt.title('gait cycle')
plt.show()
```

```python
l1=45
l2=45
```

```python
# Forward Kinematics 2R Planar Fixed Axis
def forward_kinematics(q1, q2):
    x = l1 * np.cos(q1) + l2 * np.cos(q2+q1)
    y = l1 * np.sin(q1) + l2 * np.sin(q2+q1)
    
    pos=np.array([x,y])
    print("   Position of the ankle:", pos)
    return x,y

# Inverse Kinematics 2R Planar Fixed Axis
def inverse_kinematics(x, y):
    D = (x ** 2 + y ** 2 - l1 ** 2 - l2 ** 2) / (2 * l1 * l2)
    q2 = -np.arctan2(np.sqrt(1 - D ** 2), D)
    q1 = np.arctan2(y, x) - np.arctan2(l2 * np.sin(q2), l1 + l2 * np.cos(q2))
    
    jp=np.array([np.rad2deg(q1),np.rad2deg(q2)])
    print("-->Joint parameters:", jp)
    return q1, q2
```

```python
for i in range(len(x)):
    q1,q2=inverse_kinematics(80-y[i], x[i]-25)
    forward_kinematics(q1,q2)
```
