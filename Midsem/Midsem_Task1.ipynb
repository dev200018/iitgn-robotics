```python
l1=0.25
l2=0.25
m1=0.8
m2=0.8
m3=0.8
I1=0.005
I2=0.005
I3=0.8
```

```python
def Stanford_invkin(x,y,z):
    q1=np.arctan2(y,x)
    r=np.sqrt(x**2+y**2)
    s=z-l1
    q2=np.arctan2(s,r)
    d3=np.sqrt(s**2+r**2)-l2
    
    jp=np.array([np.rad2deg(q1),np.rad2deg(q2),d3])
    print("-->Joint parameters:", jp)
    return q1,q2,d3
```

```python
def Stanford_fkin(q1,q2,d3):
    n=3
#     the wrong DH parameters that works!
#     dh=np.array([[q1, l1, 0, np.pi/2],
#                  [q2, 0, l2, 0],
#                  [0, 0, d3, 0]])

    dh=np.array([[q1, l1, 0, np.pi/2],
                 [np.pi/2+q2, 0, 0, np.pi/2],
                 [0, l2+d3, 0, 0]])
    A=[0]*n
    for i in range(n):
        theta, d, a, alpha = dh[i]
        A[i]=np.array([[np.cos(theta), -np.sin(theta)*np.cos(alpha), np.sin(theta)*np.sin(alpha), a*np.cos(theta)],
                       [np.sin(theta), np.cos(theta)*np.cos(alpha), -np.cos(theta)*np.sin(alpha), a*np.sin(theta)],
                       [0, np.sin(alpha), np.cos(alpha), d],
                       [0, 0, 0, 1]])
    ans=np.identity(4)

    for i in range(n):
        ans=ans@A[i]
    pos=ans[0:3,3]
    print("   Position of end effector:", pos)
    return pos
```

```python
#point A
q1,q2,d3=Stanford_invkin(0.45,0.075,0.1)
pos=Stanford_fkin(q1,q2,d3)
#point B
q1,q2,d3=Stanford_invkin(0.45,-0.075,0.1)
pos=Stanford_fkin(q1,q2,d3)
#point C
q1,q2,d3=Stanford_invkin(0.25,-0.075,0.1)
pos=Stanford_fkin(q1,q2,d3)
#point D
q1,q2,d3=Stanford_invkin(0.25,0.075,0.1)
pos=Stanford_fkin(q1,q2,d3)
```

```python
def dh_2_jacobian_v(dh,n,R_or_P):
    A=[0]*n
    for i in range(n):
        theta, d, a, alpha = dh[i]
        A[i]=np.array([[np.cos(theta), -np.sin(theta)*np.cos(alpha), np.sin(theta)*np.sin(alpha), a*np.cos(theta)],
                       [np.sin(theta), np.cos(theta)*np.cos(alpha), -np.cos(theta)*np.sin(alpha), a*np.sin(theta)],
                       [0, np.sin(alpha), np.cos(alpha), d],
                       [0, 0, 0, 1]])
    ans=np.identity(4)
    Z=[0]*(n+1)
    Z[0]=np.array([[0],
                   [0],
                   [1]])
    O=[0]*(n+1)
    O[0]=np.array([0, 0, 0])
    for i in range(n):
        ans=ans@A[i]
        Z[i+1]=ans[0:3, 0:3]@Z[0]
        O[i+1]=ans[0:3,3].T
    
    print("   position of end effector:", ans[0:3,3])
    
    J=[0]*n
    for i in range(n):
        if R_or_P[i]==1:
            J[i]=np.concatenate((Z[i],np.array([[0,0,0]]).T),axis=0)
        else:
            cross=np.cross(Z[i].T,(O[n]-O[i]).T)
            J[i]=np.concatenate((cross.T,Z[i]),axis=0)

    final_J=J[0]
    for i in range(1,n):
        final_J=np.concatenate((final_J,J[i]),axis=1)
    # jacobian matrix done
    Jv=final_J[0:3]
    return Jv
```

```python
def Stanford_1d(x,y,z,vx,vy,vz):
    q1,q2,d3=Stanford_invkin(x,y,z) # prints joint parameters as well
    
    n=3
    dh=np.array([[q1, l1, 0, np.pi/2],
                 [np.pi/2+q2, 0, 0, np.pi/2],
                 [0, l2+d3, 0, 0]])
    R_or_P=[0,0,1]

    Jv=dh_2_jacobian_v(dh,n,R_or_P)
    v=np.array([[vx],
                [vy],
                [vz]])
    join_velocities=np.linalg.pinv(Jv)@v
    print("   velocity of end effector:", v.T)
    print("   velocity of joint variables:", join_velocities.T)
    
def PUMA_1d(x,y,z,vx,vy,vz):
    q1,q2,q3=PUMA_invkin(x,y,z) # prints joint parameters as well
    
    n=3
    dh=np.array([[q1, l1, 0, np.pi/2],
                 [q2, 0, l2, 0],
                 [q3, 0, l3, 0]])
    R_or_P=[0,0,0]

    Jv=dh_2_jacobian_v(dh,n,R_or_P)
    v=np.array([[vx],
                [vy],
                [vz]])
    join_velocities=np.linalg.pinv(Jv)@v
    print("   velocity of end effector:", v.T)
    print("   velocity of joint variables:", join_velocities.T)
    
def SCARA_1d(x,y,z,vx,vy,vz):
    q1,q2,d3=SCARA_invkin(x,y,z) # prints joint parameters as well
    
    # finding jacobian matrix
    n=3
    dh=np.array([[q1, l1, l2, 0],
                 [q2, 0, l2, 0],
                 [0, d3, 0, 0]])
    R_or_P=[0,0,1]

    Jv=dh_2_jacobian_v(dh,n,R_or_P)
    v=np.array([[vx],
                [vy],
                [vz]])
    join_velocities=np.linalg.pinv(Jv)@v
    print("   velocity of end effector:", v.T)
    print("   velocity of joint variables:", join_velocities.T)
```

```python
def Euler_angles(R,show=1):
    if not(R[0][2]==0 and R[1][2]==0):
        q2=np.arctan2(np.sqrt(1-R[2][2]**2), R[2][2])
        q3=np.arctan2(R[2][1], -R[2][0])
        q1=np.arctan2(R[1][2], R[0][2])
    elif R[2][2]==1:
        q2=0
        q1=0
        q3=np.arctan2(R[1][0], R[0][0])
    elif R[2][2]==-1:
        q2=np.pi
        q1=0
        q3=np.arctan2(R[1][0], -R[0][0])
    else:
        print("something is wrong")
    if(show):
        print("   Euler angles:",np.around(np.rad2deg(q1),5),np.around(np.rad2deg(q2),5),np.around(np.rad2deg(q3),5))
    return q1,q2,q3

def Euler_matrix(q1,q2,q3):
    r1=np.array([[np.cos(q1), -np.sin(q1), 0],
                 [np.sin(q1), np.cos(q1), 0],
                 [0, 0, 1]])
    r2=np.array([[np.cos(q2), 0, np.sin(q2)],
                 [0, 1, 0],
                 [-np.sin(q2), 0, np.cos(q2)]])
    r3=np.array([[np.cos(q3), -np.sin(q3), 0],
                 [np.sin(q3), np.cos(q3), 0],
                 [0, 0, 1]])
    r=r1@r2@r3
    return r
```
