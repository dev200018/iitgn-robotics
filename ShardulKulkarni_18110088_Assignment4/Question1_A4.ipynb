# Inverse Position Kinematics
The function impliments the inverse position kinematics of the Stanford manipulator derived from Geometric approach

Function Input : invPos([x, y, z],[d1, d2]) (end effector co-ordinates and Link lengths)

Function Output : [theta1, theta2, d3] and No Output if Singularity

image.png

```python
def invPos(pos,linkl):

  if (pos[0]==0 and pos[1]==0) :
    print("Singularity !!!!!!!")
    return []
  else:
    theta1 = np.arctan(pos[1]/pos[0])

    r= (pos[0]**2+pos[1]**2)**(0.5)
    s=pos[2]-linkl[0]

    theta2 = np.arctan(s/r)+(np.pi/2)

    d3=(r**2+s**2)**(0.5) - linkl[1]

    return [theta1,theta2,d3]
 
```

```python
print(invPos(pos,links)) 
```
