# Inverse Orientation Kinematics
The below function implements the inverse orientation kinematics for the spherical wrist.

Function Input : Euler Rotation matrix (3 X 3)

Function Output : Euler angles in format (theta, phi, psi)

image.png

```python
def invOri_WR(u):
  if ((u[0][2]!=0 or u[1][2]!=0) and (u[2][2]==1 or u[2][2]==-1)):
    print("Not a Valid rotation matrix !!!")
    return
  
  if(u[0][2]==0 and u[1][2]==0):
    print("Infinite Solutions !!")
    return
  
  theta=np.arctan(((1-u[2][2]**2)**(0.5))/u[2][2])
  phi = np.arctan(u[1][2]/u[0][2])
  psi = np.arctan(-u[2][1]/u[2][0])

  return [theta,phi,psi]
```
