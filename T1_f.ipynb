```python
# Name - Dev Patel
# Roll No - 18110113



import numpy as np

links = 3
# Some constants
L1 = 0.25
L2 = 0.25
L3 = 0.25

A = np.array([0.4,0.06,0.15])
B = np.array([0.4,0.01,0.15])
C = np.array([0.35,0.01,0.15])
D = np.array([0.35,0.06,0.15])

R = np.array([[1,0,0],[0,1,0],[0,0,1]]) # Orientation Matrix for the wrist

#STANFORD
def inverse_kinematics_Stanford(O):

    r = (O[0]**2 + O[1]**2)**0.5
    s = O[2] - L1
    theta1 = np.arctan2(O[1],O[0]) 
    d3 = (r**2 + s**2  - L2**2)**0.5
    theta2 = -np.arctan2(-s,r) + np.arctan2(d3,L2)

    theta4 = np.arctan2(R[0,2]*np.sin(theta1)*np.cos(theta2) + R[1,2]*np.cos(theta1) + R[2,2]*np.sin(theta1)*np.sin(theta2), R[0,2]*np.cos(theta1)*np.cos(theta2) - R[1,2]*np.sin(theta1) + R[2,2]*np.cos(theta1)*np.sin(theta2))
    theta5 = np.arctan2((1 - (R[0,2]*np.sin(theta2)+R[2,2]*np.cos(theta2))**2)**0.5, R[0,2]*np.sin(theta2)+R[2,2]*np.cos(theta2))
    theta6 = np.arctan2(R[0,1]*np.sin(theta2) + R[2,1]*np.cos(theta2),-(R[0,0]*np.sin(theta2) + R[2,0]*np.cos(theta2)))

    I = np.array([theta4,theta5,theta6])
    return(I)

input = inverse_kinematics_Stanford(A)
print(" For point A \n Theta4 = ", input[0], "\n Theta5 =", input[1],"\n Theta6 ", input[2], "\n")

input = inverse_kinematics_Stanford(B)
print(" For point B \n Theta4 = ", input[0], "\n Theta5 =", input[1],"\n Theta6 ", input[2], "\n")

input = inverse_kinematics_Stanford(C)
print(" For point C \n Theta4 = ", input[0], "\n Theta5 =", input[1],"\n Theta6 ", input[2], "\n")

input = inverse_kinematics_Stanford(D)
print(" For point D \n Theta4 = ", input[0], "\n Theta5 =", input[1],"\n Theta6 ", input[2], "\n")
```

```python
# SCARA manipulator
d1 = 0.25
# Some constants
L1 = 0.25
L2 = 0.25
L3 = 0.25

R = np.array([[1,0,0],[0,1,0],[0,0,1]]) # Orientation Matrix for the wrist

def inverse_kinematics_SCARA(O):

    c2 = (O[0]**2 + O[1]**2 - L1**2 - L2**2)/(2*L1*L2)
    theta2 = np.arctan2(-(1-c2**2)**0.5,c2)
    theta1 = np.arctan2(O[1],O[0]) - np.arctan2(L2*np.sin(theta2),L1 + L2*np.cos(theta2))

    d3 = d1 - O[2] 

    theta4 = np.arctan2(np.sin(theta1+theta2)*R[0,2] - np.cos(theta1 +theta2)*R[1,2], np.cos(theta1+theta2)*R[0,2] + np.sin(theta1 +theta2)*R[1,2])
    theta5 = np.arctan2((1 - R[2,2]**2)**0.5,-R[2,2])
    theta6 = np.arctan2(-R[2,1],R[2,0])

    I = np.array([theta4,theta5,theta6])
    return(I)

input = inverse_kinematics_SCARA(A)
print(" For point A \n Theta4 = ", input[0], "\n Theta5 =", input[1],"\n Theta6 ", input[2], "\n")

input = inverse_kinematics_SCARA(B)
print(" For point B \n Theta4 = ", input[0], "\n Theta5 =", input[1],"\n Theta6 ", input[2], "\n")

input = inverse_kinematics_SCARA(C)
print(" For point C \n Theta4 = ", input[0], "\n Theta5 =", input[1],"\n Theta6 ", input[2], "\n")

input = inverse_kinematics_SCARA(D)
print(" For point D \n Theta4 = ", input[0], "\n Theta5 =", input[1],"\n Theta6 ", input[2], "\n")
```

```python
# PUMA

R = np.array([[1,0,0],[0,1,0],[0,0,1]]) # Orientation Matrix for the wrist

def inverse_kinematics_PUMA(O):

    r = (O[0]**2 + O[1]**2)**0.5
    s = O[2] - L1
    theta1 = np.arctan2(O[1],O[0])
    d = (r**2 + s**2 - L2**2 - L3**2)/(2*L2*L3)
    theta3 = np.arctan2(((1-d**2)**0.5),d)
    theta2 = np.arctan2(s,r) - np.arctan2(L3*np.sin(theta3),L2 + L3*np.cos(theta3))

    theta4 = np.arctan2(-np.cos(theta1)*np.sin(theta2 + theta3)*R[0,2] -np.sin(theta1)*np.sin(theta2 + theta3)*R[1,2] + np.cos(theta2 + theta3)*R[2,2], np.cos(theta1)*np.cos(theta2 + theta3)*R[0,2] + np.sin(theta1)*np.cos(theta2 + theta3)*R[1,2] + np.sin(theta2 + theta3)*R[2,2])
    theta5 = np.arctan2((1 - (np.sin(theta1)*R[0,2] - np.cos(theta1)*R[1,2])**2)**0.5, np.sin(theta1)*R[0,2] - np.cos(theta1)*R[1,2])
    theta6 = np.arctan2(np.sin(theta1)*R[0,1] - np.cos(theta1)*R[1,1], -np.sin(theta1)*R[0,0] + np.cos(theta1)*R[1,0])

    I = np.array([theta4,theta5,theta6])
    return(I)

input = inverse_kinematics_PUMA(A)
print(" For point A \n Theta4 = ", input[0], "\n Theta5 =", input[1],"\n Theta6 ", input[2], "\n")

input = inverse_kinematics_PUMA(B)
print(" For point B \n Theta4 = ", input[0], "\n Theta5 =", input[1],"\n Theta6 ", input[2], "\n")

input = inverse_kinematics_PUMA(C)
print(" For point C \n Theta4 = ", input[0], "\n Theta5 =", input[1],"\n Theta6 ", input[2], "\n")

input = inverse_kinematics_PUMA(D)
print(" For point D \n Theta4 = ", input[0], "\n Theta5 =", input[1],"\n Theta6 ", input[2], "\n")
```
