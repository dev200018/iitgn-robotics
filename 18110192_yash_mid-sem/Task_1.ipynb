```python
import math
import numpy as np

def deg(theta):
    rad = (math.pi/180)*theta
    return rad

def cos(q):
    return math.cos(deg(q))

def sin(q):
    return math.sin(deg(q))
```

# Robot 1 - Stanford-type (RRP)

```python
# Taken from the repository and modified a bit
def standford_inverse_kinematics(endeffector_position,lengthsoflinks):
    theta1 = np.arctan(endeffector_position[1]/endeffector_position[0])
    r = np.sqrt(endeffector_position[0]**2 + endeffector_position[1]**2)
    s = endeffector_position[2] - lengthsoflinks[0]
    theta2 = np.arctan(s/r)
    d3 = np.sqrt(r**2 + s**2) - lengthsoflinks[1]
    theta1 = theta1*180/np.pi
    theta2 = theta2*180/np.pi
    print(" Theta1 = ", theta1, " deg.\n Theta2 =", theta2," deg.\n Extension: ", d3, "\n")
    return [theta1, theta2, d3]
```

```python
def R_(ap, teta):
    R = np.array([[cos(teta), -sin(teta)*cos(ap),  sin(teta)*sin(ap)],
                  [sin(teta),  cos(teta)*cos(ap), -cos(teta)*sin(ap)],
                  [        0,            sin(ap),            cos(ap)]])
    return R
def P_(a, d, teta):
    P = np.array([[a*cos(teta)],
                  [a*sin(teta)],
                  [d]])
    return P

def forward_kinematics(n, DH):
    T = np.identity(4, dtype = float)
    extra_row = np.array([0, 0, 0, 1])

    for i in range(n):
        a_i = DH[i][0]
        ap_i = DH[i][1]
        d_i = DH[i][2]
        teta_i = DH[i][3]

        R_i = R_(ap_i, teta_i)
        P_i = P_(a_i, d_i, teta_i)
        A_i = np.vstack((np.concatenate((R_i, P_i), axis=1), extra_row))
        T = np.dot(T, A_i)

    Pn = T[:3,3]
#     print('T = \n', T, '\n')
    print(' Estimated End-effector position = ', Pn, '\n')
```

```python
print("For Robot 1 - Standford-type (RRP) \n##################################################\n")
pos = [A, B, C, D]
name = ['A', 'B', 'C', 'D']
for i in range(len(pos)):
    print("For point ", name[i], '\n')
    j_p = standford_inverse_kinematics(pos[i],l)
    n = 3                                               # Number of links
    DH = np.array([[0, 90, l1, j_p[0]],                 # DH Parameters (EACH ROW IN ORDER [ai, alpha_i, di, theta_i])
                   [l2, 0, 0, j_p[1]],
                   [j_p[2], 0, 0, 0]])
    forward_kinematics(n, DH)
    print(' Actual End-effector position = ', pos[i], '\n \n********************************************************************************************************* \n')
```

# Robot 2 - PUMA-type (RRR)

```python
def puma_inverse_kinematics(endeffector_position,lengthsoflinks):
    x = endeffector_position[0]
    y = endeffector_position[1]
    z = endeffector_position[2]
    L1 = lengthsoflinks[0]
    L2 = lengthsoflinks[1]
    L3 = lengthsoflinks[2]
    theta1 = np.arctan(y/x)*180/np.pi
    x = x/cos(theta1)
    y = y/sin(theta1)
    z = z - L1    
    l1 = L2
    l2 = L3
    y = z
    theta = math.acos((x**2 + y**2 - l1**2 - l2**2)/(2*l1*l2)) * 180/math.pi
    if (x>=0 and y>=0):
        q1 = (math.atan(abs(y)/abs(x)) - math.atan((l2*sin(theta))/(l1 + l2*cos(theta)))) * 180/math.pi
    elif (x<=0 and y>=0):
        q1 = (math.pi - math.atan(abs(y)/abs(x)) - math.atan((l2*sin(theta))/(l1 + l2*cos(theta)))) * 180/math.pi
    elif (x<=0 and y<=0):
        q1 = (math.pi + math.atan(abs(y)/abs(x)) - math.atan((l2*sin(theta))/(l1 + l2*cos(theta)))) * 180/math.pi
    else:
        q1 = (2*math.pi - math.atan(abs(y)/abs(x)) - math.atan((l2*sin(theta))/(l1 + l2*cos(theta)))) * 180/math.pi

    q2 = theta
    theta2 = q1
    theta3 = q2
    
    print(" Theta1 = ", theta1, " deg.\n Theta2 =", theta2," deg.\n Theta3: ", theta3, "\n")
    return [theta1, theta2, theta3]
```

```python
print("For Robot 2 - PUMA-type (RRR) \n##################################################\n")
pos = [A, B, C, D]
name = ['A', 'B', 'C', 'D']
for i in range(len(pos)):
    print("For point ", name[i], '\n')
    j_p = puma_inverse_kinematics(pos[i],l)
    n = 3                                               # Number of links
    DH = np.array([[0, 90, l1, j_p[0]],                 # DH Parameters (EACH ROW IN ORDER [ai, alpha_i, di, theta_i])
                   [l2, 0, 0, j_p[1]],
                   [l3, 0, 0, j_p[2]]])
    forward_kinematics(n, DH)
    print(' Actual End-effector position = ', pos[i], '\n \n********************************************************************************************************* \n')
```

# Robot 3 - SCARA-type (RRP)

```python
def scara_inverse_kinematics(endeffector_position,lengthsoflinks):
    x = endeffector_position[0]
    y = endeffector_position[1]
    z = endeffector_position[2]
    l1 = lengthsoflinks[0]
    l2 = lengthsoflinks[1]
    theta = math.acos((x**2 + y**2 - l1**2 - l2**2)/(2*l1*l2)) * 180/math.pi
    if (x>=0 and y>=0):
        q1 = (math.atan(abs(y)/abs(x)) - math.atan((l2*sin(theta))/(l1 + l2*cos(theta)))) * 180/math.pi
    elif (x<=0 and y>=0):
        q1 = (math.pi - math.atan(abs(y)/abs(x)) - math.atan((l2*sin(theta))/(l1 + l2*cos(theta)))) * 180/math.pi
    elif (x<=0 and y<=0):
        q1 = (math.pi + math.atan(abs(y)/abs(x)) - math.atan((l2*sin(theta))/(l1 + l2*cos(theta)))) * 180/math.pi
    else:
        q1 = (2*math.pi - math.atan(abs(y)/abs(x)) - math.atan((l2*sin(theta))/(l1 + l2*cos(theta)))) * 180/math.pi

    q2 = theta
    theta1 = q1
    theta2 = q2
    d3 = -z
    print(" Theta1 = ", theta1, " deg.\n Theta2 =", theta2," deg.\n Extension: ", d3, "\n")
    return [theta1, theta2, d3]
```

```python
print("For Robot 3 - SCARA-type (RRP) \n##################################################\n")
pos = [A, B, C, D]
name = ['A', 'B', 'C', 'D']
for i in range(len(pos)):
    print("For point ", name[i],'\n')
    j_p = scara_inverse_kinematics(pos[i],l)
    n = 3                                              # Number of links
    DH = np.array([[l1, 0, 0, j_p[0]],                 # DH Parameters (EACH ROW IN ORDER [ai, alpha_i, di, theta_i])
                   [l2, 180, 0, j_p[1]],
                   [0, 0, j_p[2], 0]])
    forward_kinematics(n, DH)
    print(' Actual End-effector position = ', pos[i], '\n \n********************************************************************************************************* \n')
```
