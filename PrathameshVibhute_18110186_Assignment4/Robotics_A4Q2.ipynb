```python
x=float(input('x='))
y=float(input('y='))
z=float(input('z='))
position=[x,y,z]
d1=float(input('d1='))
d2=float(input('d2='))
d3=float(input('d3='))
linklengths=[d1,d2,d3]
```

Defining function inverse to calculate inverse position kinematics

```python
def inv_scara(linklen,pos):
  r=((pos[0]**2+pos[1]**2-linklen[0]**2-linklen[1]**2)/(2*linklen[0]*linklen[1]))**(1/2)
  theta2=np.arctan([r/(1-r**2)**(1/2),-(r/(1-r**2)**(1/2))])
  theta1=np.arctan(pos[1]/pos[0])-np.arctan(linklen[1]*np.sin(theta2)/(linklen[0]+linklen[1]*np.cos(theta2)))
  d4=linklen[2]-pos[2]

  return [theta1,theta2,d4]
```

```python
inv_scara(linklengths,position)
```
