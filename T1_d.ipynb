```python
# Name - Dev Patel
# Roll No - 18110113



import numpy as np

# Some constants
L1 = 0.25
L2 = 0.25
L3 = 0.25


# Get a skew symmetric matrix for a given 3x1 matrix
def skew(x):
    return np.array([[0, -x[2], x[1]],
                     [x[2], 0, -x[0]],
                     [-x[1], x[0], 0]])

#STANFORD
def inverse_kinematics_Stanford(O):

    r = (O[0]**2 + O[1]**2)**0.5
    s = O[2] - L1
    theta1 = np.arctan2(O[1],O[0]) 
    d3 = (r**2 + s**2  - L2**2)**0.5
    theta2 = -np.arctan2(-s,r) + np.arctan2(d3,L2)
    

    I = np.array([theta1,theta2,d3])
    return(I)

# Get the jacobian matrix for a robot
def Jacobian_Stanford(DH_parameters):
    T = np.eye(4)
    O = np.array([0,0,0])
    Z = np.array([0,0,1])
    for i in range(4):
        A = np.array([[np.cos(DH_parameters[i,0]), -np.sin(DH_parameters[i,0])*np.cos(DH_parameters[i,3]), np.sin(DH_parameters[i,0])*np.sin(DH_parameters[i,3]),DH_parameters[i,1]*np.cos(DH_parameters[i,0])],\
                [np.sin(DH_parameters[i,0]), np.cos(DH_parameters[i,0])*np.cos(DH_parameters[i,3]), -np.cos(DH_parameters[i,0])*np.sin(DH_parameters[i,3]),DH_parameters[i,1]*np.sin(DH_parameters[i,0])],\
                    [0,np.sin(DH_parameters[i,3]),np.cos(DH_parameters[i,3]),DH_parameters[i,2]],\
                        [0,0,0,1]])

        T = np.dot(T,A)
        O = np.c_[O,np.transpose(T[:3,3])]
        Z = np.c_[Z,np.transpose(T[:3,2])]
    J = np.r_[np.dot(skew(Z[:,0]),np.transpose(O[:,3]) - np.transpose(O[:,0])), np.transpose(Z[:,0])]
    for i in range(1,3):
        if(i==3):
             J = np.c_[J,np.r_[np.transpose(Z[:,i]),np.transpose([0,0,0])]]
        else:
            J = np.c_[J,np.r_[np.dot(skew(Z[:,i]),np.transpose(O[:,3]) - np.transpose(O[:,i])), np.transpose(Z[:,i])]]
    return(J)

def check_end_Effector_position_Stanford(I):
    T = np.eye(4)
    DH_parameters = np.array([[I[0],0,L1,1.57],[I[1],L2,0,1.57],[0,0,I[2],0]])
    for i in range(4):
        A = np.array([[np.cos(DH_parameters[i,0]), -np.sin(DH_parameters[i,0])*np.cos(DH_parameters[i,3]), np.sin(DH_parameters[i,0])*np.sin(DH_parameters[i,3]),DH_parameters[i,1]*np.cos(DH_parameters[i,0])],\
                [np.sin(DH_parameters[i,0]), np.cos(DH_parameters[i,0])*np.cos(DH_parameters[i,3]), -np.cos(DH_parameters[i,0])*np.sin(DH_parameters[i,3]),DH_parameters[i,1]*np.sin(DH_parameters[i,0])],\
                    [0,np.sin(DH_parameters[i,3]),np.cos(DH_parameters[i,3]),DH_parameters[i,2]],\
                        [0,0,0,1]])

        T = np.dot(T,A)
    return(print('x=',T[0,3],'y=',T[1,3],'z=',T[2,3]))

# Motion A-B
V = np.array([0,-0.01,0,0,0,0])
for i in range(4):
    P = np.array([0.4,0.05 - 0.01*i,0.1])
    I = inverse_kinematics_Stanford(P)
    DH = np.array([[I[0],0,L1,1.57],[I[1],L2,0,1.57],[0,0,I[2],0]])
    J = Jacobian_Stanford(DH)
    q_dot = np.matmul(np.linalg.pinv(J),np.transpose(V))
    print("\n At point: ",P,"\n joint velocities are: ", q_dot[:])
    print(" Joint Variables: ",I,"\n Checking:")
    Check = print(check_end_Effector_position_Stanford(I))
    CH = print(np.dot(J,q_dot))

# Motion B-C
V = np.array([-0.01,0,0,0,0,0])
for i in range(4):
    P = np.array([0.4-0.01*i,0.01,0.1])
    I = inverse_kinematics_Stanford(P)
    DH = np.array([[I[0],0,L1,1.57],[I[1],L2,0,1.57],[0,0,I[2],0]])
    J = Jacobian_Stanford(DH)
    q_dot = np.matmul(np.linalg.pinv(J),np.transpose(V))
    print("\n At point: ",P,"\n joint velocities are: ", q_dot[:])
    print(" Joint Variables: ",I,"\n Checking:")
    Check = print(check_end_Effector_position_Stanford(I))
    CH = print(np.dot(J,q_dot))

# Motion C-D
V = np.array([0,0.01,0,0,0,0])
for i in range(4):
    P = np.array([0.35,0.01 + 0.01*i,0.1])
    I = inverse_kinematics_Stanford(P)
    DH = np.array([[I[0],0,L1,1.57],[I[1],L2,0,1.57],[0,0,I[2],0]])
    J = Jacobian_Stanford(DH)
    q_dot = np.matmul(np.linalg.pinv(J),np.transpose(V))
    print("\n At point: ",P,"\n joint velocities are: ", q_dot[:])
    print(" Joint Variables: ",I,"\n Checking:")
    Check = print(check_end_Effector_position_Stanford(I))
    CH = print(np.dot(J,q_dot))

# Motion D-A
V = np.array([0.01,0,0,0,0,0])
for i in range(4):
    P = np.array([0.35+0.01*i,0.05,0.1])
    I = inverse_kinematics_Stanford(P)
    DH = np.array([[I[0],0,L1,1.57],[I[1],L2,0,1.57],[0,0,I[2],0]])
    J = Jacobian_Stanford(DH)
    q_dot = np.matmul(np.linalg.pinv(J),np.transpose(V))
    print("\n At point: ",P,"\n joint velocities are: ", q_dot)
    print(" Joint Variables: ",I,"\n Checking:")
    Check = print(check_end_Effector_position_Stanford(I))
    CH = print(np.dot(J,q_dot))
```
