```python
import xlrd
import matplotlib.pyplot as plt
```

# Gait Trajectory Plot
To run the code, the file 'Gait_DATA.xlsx' should be in the same directory as the code file.

```python
loc = ("./Gait_DATA.xlsx")
 
wb = xlrd.open_workbook(loc)
sheet = wb.sheet_by_index(0)
x=[]
y=[]
for i in range(1,sheet.nrows):
  x.append(sheet.cell_value(i,0))
  y.append(sheet.cell_value(i,1))

plt.plot(x,y)
plt.title("Gait Trajectory")
plt.xlabel("X (cm)")
plt.ylabel("Y (cm)")
plt.show()
```
